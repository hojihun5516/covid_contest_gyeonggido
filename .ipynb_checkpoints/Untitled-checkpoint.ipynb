```python
people = sorted(people.items())
people
city_df = pd.DataFrame(people)
city_df = city_df.rename(columns={0:'city', 1:'total'})
# city_index = list(city_df['city'])
# city_index
# city_df = city_df.reindex(city_index)
city_df.head()
```

```python
city_infected_list = []
city_infected = dict(covid_df['지역 + 시 '].value_counts())
city_infected = sorted(city_infected.items())
city_infected = pd.Series(city_infected)
for v in city_infected:
    city_infected_list.append(v[1])
```

```python
city_df = pd.concat([city_df, pd.Series(city_infected_list)],  axis=1)
city_df = city_df.rename(columns={0:'infected'})
city_df
```

```python
city_df['infected_percentage'] = city_df['infected'] / city_df['total']*100
city_df = city_df.set_index('city')
city_df
```

# 제공 COVID_Data

## 제거 목록
- '기준일(발병일, 확진일 선택)' 제거 => 확진일자와 동일
- 지역 (group) 제거 => "지역 + 시 " 와동일
- 무증상/조사중 제거 => 무증상자 데이터는 무증상자수에 있기때문에 제거 
- 필터(무증상,조사중 제외) 제거 => 필터(무증상,조사중 제외) 필요없는 제거
- 경기번호 제거 => 사람의 구분 고유 아이디임
- 나이(만) 제거 => 연령대가 있기 때문
- 무증상/조사중 기준일 제거 => 불필요한 데이터라 판단
- 연번 제거 => 사람의 고유 값 
- 증상발현일 제거 => 무의미한 데이터 "확진일-증상발현일_무증상조사중포함" 있음
- 필터(무증상,조사중 제외)(복사본) 제거 => All True로 무의미
- 지역 제거 => 지역 + 시 를 쪼개놓음
- 확진일-증상발현일 제거 => 확진일-증상발현일_무증상조사중포함과 같음 남겨 놓은 데이터가 표시가 잘 되어있음
- 확진자 제거 => 환자 고유아이디
- 레코드 수 제거 => All True 무의미
- 최종_무증상자수 제거 => 무증상자수와 같음
