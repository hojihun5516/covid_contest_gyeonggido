#### 'Age'의 Nan 값을 대체하기 위해 호칭 title 분리

```python
age_nan = data_df[data_df['Age'].isnull()]
age_nan.tail()
```

```python
age_nan = data_df[data_df['Age'].isnull()]
title = [re.split('. |,', i.split(', ')[1])[0] for i in age_nan['Name']]
age_nan['Title'] = title
age_nan
```

```python
set(title)
```

```python
data_df_withtitle = data_df.copy()
```

```python
# age_nan = data_df[data_df['Age'].isnull()]
title = [re.split('. |,', i.split(', ')[1])[0] for i in data_df_withtitle['Name']]
data_df_withtitle['Title'] = title
data_df_withtitle.head(2)
```

```python
age_mean_dr = data_df_withtitle[data_df_withtitle['Title']=='Dr']['Age'].mean()
age_mean_master = data_df_withtitle[data_df_withtitle['Title']=='Master']['Age'].mean()
age_mean_miss = data_df_withtitle[data_df_withtitle['Title']=='Miss']['Age'].mean()
age_mean_mr = data_df_withtitle[data_df_withtitle['Title']=='Mr']['Age'].mean()
age_mean_mrs = data_df_withtitle[data_df_withtitle['Title']=='Mrs']['Age'].mean()
```

```python
for i in age_nan['PassengerId'].index:
    title = data_df_withtitle.loc[i]['Title']
    if title == 'Dr':
        data_df_withtitle.at[i, 'Age'] = age_mean_dr
    elif title == 'Master':
        data_df_withtitle.at[i, 'Age'] = age_mean_master
    elif title == 'Miss':
        data_df_withtitle.at[i, 'Age'] = age_mean_miss
    elif title == 'Mr':
        data_df_withtitle.at[i, 'Age'] = age_mean_mr
    elif title == 'Mrs':
        data_df_withtitle.at[i, 'Age'] = age_mean_mrs
```

```python
data_df_withtitle['Age'] = data_df_withtitle['Age'].apply(lambda x : int(np.round(x)))
```

##  Train - Test split

```python
y_data = data_df_withtitle[['Survived']]
y = np.array(y_data)

del data_df_withtitle['Survived']
x_data = data_df_withtitle.copy()
X = np.array(x_data)
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Gradient Boosting classification

```python
params = {'n_estimators': 1000,
          'max_depth': 3,
          'min_samples_split': 5,
          'learning_rate': 0.01}
```

```python
clf = ensemble.GradientBoostingClassifier(**params)
clf.fit(x_train, y_train)

acc = accuracy_score(y_test, clf.predict(x_test))
print("The accuracy score on test set: {:.4f}".format(acc))
```

```python
train_score = np.zeros((params['n_estimators'],), dtype=np.float64)
for i, y_pred in enumerate(clf.staged_predict(x_train)):
    train_score[i] = accuracy_score(y_train, y_pred)

test_score = np.zeros((params['n_estimators'],), dtype=np.float64)
for i, y_pred in enumerate(clf.staged_predict(x_test)):
    test_score[i] = accuracy_score(y_test, y_pred)
    
fig = plt.figure(figsize=(12, 6))
plt.subplot(1, 1, 1)
plt.title('Accuracy') # Binomial deviance loss function for binary classification
plt.plot(np.arange(params['n_estimators']) + 1, train_score, 'b-', label='Training Set Accuracy')
plt.plot(np.arange(params['n_estimators']) + 1, test_score, 'r-', label='Test Set Accuracy')
plt.legend(loc='upper right')
plt.xlabel('Boosting Iterations')
plt.ylabel('Accuracy')
fig.tight_layout()
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

fpr, tpr, _ = roc_curve(y_true=y_test, y_score=clf.predict_proba(x_test)[:,1]) # real y & predicted y (based on "Sepal width")
roc_auc = auc(fpr, tpr) # AUC 면적의 값 (수치)

plt.figure(figsize=(10, 10))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.title("ROC curve")
plt.show()
```

```python
from sklearn.metrics import classification_report

predictions = clf.predict(x_test)
print(classification_report(y_test, predictions)) 
print("Accuracy on Training set: {:.3f}".format(clf.score(x_train, y_train)))
print("Accuracy on Test set: {:.3f}".format(clf.score(x_test, y_test)))
```

```python
for index in data_list:
    print('[[', index,']]')
#     X = np.array(data_all)
    X = np.array(globals()['{}'.format(index)])
    y = np.array(y_data)

    x_train, x_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=0)


    ### KNeighborsClassifier
    print('KNeighborsClassifier')
    temp_i = 0
    temp_a = 0
    for i in [3, 6, 9, 12, 20, 30]:
        knn = neighbors.KNeighborsClassifier(n_neighbors=i)
        knn.fit(x_train, y_train)
        knn_pred_test = knn.predict(x_test)
#         print('neighbors: {},  Accuracy: {}'.format(i, accuracy_score(knn_pred_test, y_test)))
        if temp_a < accuracy_score(knn_pred_test, y_test):
            temp_i = i
            temp_a = accuracy_score(knn_pred_test, y_test)
    print('neighbors: {},  Accuracy: {}'.format(temp_i, temp_a))
    print('\n')
```

## SVC

```python
print('SVC')
for index in data_list:
    X = np.array(globals()['{}'.format(index)])
    y = np.array(y)
    x_train, x_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=0)
    
    print(index)
    # 아래 param_grid dict 의 C & gamma 에 후보 Hyper-params 값들을 리스트업합니다.
    param_grid = {'C' : [0.1, 1, 10, 100, 1000], 
                 'gamma' : [1, 0.1, 0.01, 0.001, 0.0001],
                 'kernel' : ['rbf']}

#     print('GridSearchCV')
    grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=0)
    # refit : 찾아진 가장 좋은 params로 estimator를 setting할 지 여부 (setting해줘야 곧바로 predict가 가능)
    # verbose : 설명의 자세한 정도 (verbose를 3과 같이 바꿔보시면 더 자세하게 매 param set 마다의 결과를 확인할 수 있습니다.)
    grid.fit(x_train, y_train)
    print('The best parameters are ', grid.best_params_)
    print()

    ### SVC
#     print('SVC')
    svm = SVC(kernel='rbf', C=grid.best_params_['C'], gamma=grid.best_params_['gamma']).fit(x_train, y_train) 
    # svm.fit(x_train, y_train)

    print("Accuracy on Training set: {:.3f}".format(svm.score(x_train, y_train)))
    print("Accuracy on Test set: {:.3f}".format(svm.score(x_test, y_test)))
    print()
    print()
```
